```python
class Data_Visualizer:
    
    def __init__(self, parent_dir=None):
        super(Data_Visualizer,self).__init__()
        self.parent_dir= parent_dir
        
    def get_files(self):
        self.files= []
        
        for (_,_,filenames) in walk(self.parent_dir):
            self.files.append(filenames)
        self.files= np.vstack(np.array(self.files))
        self.files=np.reshape(self.files,(self.files.shape[1],))
        
        return self.files
    
    def load_file(self, file):
        data= np.load(file, allow_pickle=True)
        temp= data.item()['x_train']
        label= data.item()['labels']
        temp= temp.reshape((temp.shape[0], 28,28))
        return temp, label
    
    def visualize(self):
        self.get_files()
        for file in self.files:
            if "train" in file:
                
                file= os.path.join(self.parent_dir,file)
                data,label= self.load_file(file)
                print(f"Visualizing task for the class {label} and {file}")
                plt.imshow(data[0])
                plt.show()
        
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tqdm import tqdm

num_classes = 10
input_shape = (28, 28, 1)
n_residual_blocks = 5
# The data, split between train and test sets
(x, _), (y, _) = keras.datasets.mnist.load_data()
# Concatenate all of the images together
data = np.concatenate((x, y), axis=0)
# Round all pixel values less than 33% of the max 256 value to 0
# anything above this value gets rounded up to 1 so that all values are either
# 0 or 1
data = np.where(data < (0.33 * 256), 0, 1)
data = data.astype(np.float32)

class PixelConvLayer(layers.Layer):
    def __init__(self, mask_type, **kwargs):
        super(PixelConvLayer, self).__init__()
        self.mask_type = mask_type
        self.conv = layers.Conv2D(**kwargs)

    def build(self, input_shape):
        # Build the conv2d layer to initialize kernel variables
        self.conv.build(input_shape)
        # Use the initialized kernel to create the mask
        kernel_shape = self.conv.kernel.get_shape()
        print("shape of kernel:", kernel_shape)
        self.mask = np.zeros(shape=kernel_shape)
        self.mask[: kernel_shape[0] // 2, ...] = 1.0
        print(self.mask[2,4,...])
        self.mask[kernel_shape[0] // 2, : kernel_shape[1] // 2, ...] = 1.0
        
        if self.mask_type == "B":
            self.mask[kernel_shape[0] // 2, kernel_shape[1] // 2, ...] = 1.0

    def call(self, inputs):
        self.conv.kernel.assign(self.conv.kernel * self.mask)
        return self.conv(inputs)
```
